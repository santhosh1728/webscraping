## web scraping **is an automated method used extra large amount units of data from websites. The data on the websites are unstructered. Web scraping helps to colect these unstructred data and store in a structured form***






```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
jumia = requests.get('https://www.jumia.co.ke/coke/')
```

```python
jumia
```

```python
jumia.content
```

```python
name_info = []
price_info = []
Rating_info = []
```

```python
for page in range(1,51):
  url = "https://www.jumia.co.ke/coke/" + "?page" + str(page) + " #catalog.listing"
  furl = requests.get(url)
  jsoup = BeautifulSoup(furl.content, 'html.parser')
  products = jsoup.find_all('div',class_ = 'info')

  for product in products:

    Name = product.find('h3', class_="name").text.replace('\n' , ' ')
    Price = product.find('div',class_="prc").text.replace('\n', ' ')

    try:
      Rating = product.find('div',class_='start_s').text.replace('\n', ' ')
    except:
      Rating = 'None'

    name_info.append(Name)
    price_info.append(Price)
    Rating_info.append(Rating)
```

```python
dic = {'Product_name':name_info, 'Price':price_info,'Rating':Rating_info}
```

```python
df = pd.DataFrame(dic)
```

```python
df
```

```python
df.to_csv('product_from_jomia' , index ='False', encoding  = 'utf-8' )
```
